```python
import yfinance as yf
import pandas as pd
```

```python
ms = yf.Ticker('MSFT')
```

```python
ms.history(period='max')['Close'].plot()
```

```python
ap = yf.Ticker('AAPL')
ap.history(period='max')['Close'].plot()
```

```python
gog = yf.Ticker('GOOG')
gog.history(period='max')['Close'].plot()
```

```python
ibm = yf.Ticker('IBM')
ibm.history(period='max')['Close'].plot()
```

```python
import matplotlib.pyplot as plt
```

```python
dell = yf.Ticker('DELL')
```

```python
fig,ax=plt.subplots()
ax.plot(ms.history(period='max')['Close'],label='Microsoft')
ax.plot(ap.history(period='max')['Close'],label='Apple')

ax.plot(ibm.history(period='max')['Close'],label='IBM')

plt.legend()
```

```python
fig,ax=plt.subplots()
ax.plot(gog.history(period='max')['Close'],label='Google')
ax.plot(dell.history(period='max')['Close'],label='DELL')
plt.legend()
```

```python
amd=yf.Ticker('AMD')
intel=yf.Ticker('INTC')
nvidia=yf.Ticker('NVDA')
fig,ax=plt.subplots()
ax.plot(amd.history(period='max')['Close'],label='AMD')
ax.plot(intel.history(period='max')['Close'],label='Intel')
ax.plot(nvidia.history(period='max')['Close'],label='Nvidia')
plt.legend()
```
