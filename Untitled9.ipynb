```python
def interes(P,k,t):
  return P*(1+k*t)
```

```python
interes(10000,0.04,3)
```

```python
montos=[10000]
for año in range(15):
  montos.append(round(interes(montos[año],0.04,año+1),2))
```

```python
round(montos[3],2)
```

```python
montos
```
